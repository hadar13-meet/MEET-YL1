```python
import csv
```

```python
f = open('melanoma.csv', 'rb')
reader = csv.reader(f)

year = []
incidence = []
index = 0
for row in reader:
    if index != 0:
        a = row
        r = map(float, a) 
        year.append(r[0])
        incidence.append(r[1])
    index += 1
print year 
print incidence


f.close()
```

```python
from pylab import *
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Melanoma incidences over years')
ax.set_xlabel('Year')
ax.set_ylabel('Number of incidences')
plot(year, incidence)
```

This plot shows the number of melanoma incidences over the years. 
You can see that there is no clear correlation between every single year and the amount of incidences (there are some years where the number goes up and some where it goes down), but overall, you can see that over the years, there are more and more incidences.
